```python
SEED = 2222
EARLY_STOP_NUM = 8
```

```python
# Define dependent and independent variables
X = numeric_columns.drop(columns=['AESEVCD','AESOC_1','AESOC_2','AESOC_3','AESOC_4','AESOC_5','AESOC_6','AESOC_7','AESOC_8','AESOC_9','AESOC_10','AESOC_11','Duration'])
y = numeric_columns['AESOC_1']
```
